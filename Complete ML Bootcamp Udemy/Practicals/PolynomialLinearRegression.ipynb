```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating dataset for poluynomial regression
X = 6 * np.random.rand(100,1) - 3
y = 0.5 * X**2 + 1.5 * X + np.random.randn(100,1)

# creating polynomial eqn 0.5X^2 + 1.5X + outlier
```

```python
plt.scatter(X,y,color = 'b')

plt.xlabel('X axis')
plt.ylabel('y axis')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg  = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test,y_pred)
```

```python
r2
```

```python
plt.plot(X_train,reg.predict(X_train),'k')
plt.scatter(X_train,y_train)

```

#### Observation

##### Since the dataset is quadratics linear regression wont be much help iof there is curve in the best fit line line we can reduce error

```python
# using polynomial features

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
reg.fit(X_train_poly,y_train)
y_pred = reg.predict(X_test_poly)
r2 = r2_score(y_test,y_pred)
r2
```
