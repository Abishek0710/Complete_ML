```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd

df = pd.DataFrame(dataset.data, columns=dataset['feature_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
X = df
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## Hyper parametric tuning

```python
param = {
    'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5,10,15,20,25],
    'max_features' : ['sqrt' , 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=regressor, param_grid=param, cv =5, scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

## Visualizing the tree for best parameter

```python
best_model = DecisionTreeRegressor(criterion='friedman_mse', max_depth=3,max_features='log2', splitter='best')
```

```python
best_model.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(best_model, filled= True)
```
