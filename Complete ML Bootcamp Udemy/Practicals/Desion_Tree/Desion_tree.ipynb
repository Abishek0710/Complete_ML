```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
iris['target']
```

```python
X = pd.DataFrame(iris['data'],columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
```

```python
X.head()
```

```python
y = iris['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
#apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
## visualizing decision tree

from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
# post pruning the model
# Since after max deth 2 there is not much difference in the values
# If we build the decision tree till end it will lead to overfitting
model1 = DecisionTreeClassifier(max_depth=2)
model1.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model1,filled=True)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Prepruning the model

```python
param = {
    'criterion' : ['gini','entropy','log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5], 
    'max_features' : ['sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(estimator=treemodel,param_grid=param,cv=5, scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
